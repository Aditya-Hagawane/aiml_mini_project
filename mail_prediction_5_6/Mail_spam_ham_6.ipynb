# 🔹 Dataset Explanation

The dataset contains email messages labeled as 'spam' or 'ham'.

- Spam: Unwanted or promotional messages
- Ham: Normal messages

The dataset was slightly imbalanced, with more ham messages than spam messages.

# 🔹 Model Building Steps

1. Data Cleaning:
   - Removed duplicate entries
   - Converted labels (spam = 1, ham = 0)

2. Text Processing:
   - Converted text into numerical form using TF-IDF Vectorizer

3. Data Splitting:
   - Split data into training and testing sets

4. Model Training:
   - Used Multinomial Naive Bayes algorithm

5. Evaluation:
   - Evaluated using Accuracy, Precision, Recall, and Confusion Matrix

# 🔹 Model Performance


The model achieved an accuracy of approximately 96%.

It showed high precision for spam detection, meaning it rarely misclassifies normal emails as spam.
However, recall was slightly lower, indicating that some spam messages were not detected.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
mail_data['Category'] = mail_data['Category'].map({'spam':1,'ham':0})
```

```python
X = mail_data['Message']
y = mail_data['Category']
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(mail_data['Message'])
y = mail_data['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
naive_model = MultinomialNB()
naive_model.fit(X_train, y_train)
y_pred = naive_model.predict(X_test)
```

```python
print("Accuracy Score :",accuracy_score(y_test, y_pred))
print("confusion_matrix :\n",confusion_matrix(y_test,y_pred))
print("classification report :\n",classification_report(y_test, y_pred))
```

```python
input_mail = ["URGENT! You have won $1000 cash. Click now to claim prize"]
input_data = vectorizer.transform(input_mail)

prediction = naive_model.predict(input_data)

if prediction[0] == 1:
    print("Spam Mail")
else:
    print("Not Spam Mail")
```

```python
input_mail = ["Please send me the report by evening."]
input_data = vectorizer.transform(input_mail)

prediction = naive_model.predict(input_data)

if prediction[0] == 1:
    print("Spam Mail")
else:
    print("Not Spam Mail")
```
